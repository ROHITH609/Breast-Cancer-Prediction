# BREAST CANCER DETECTION

### IMPORT REQUIRED LIBRARY

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
import warnings
warnings.simplefilter('ignore')
```

```python
dataset=sklearn.datasets.load_breast_cancer()
```

### READING DATASET

```python
dataset
```

### CONVERTING INTO DATAFRAME

```python
breast1=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
breast1
```

```python
breast1['label']=dataset.target
```

```python
breast1
```

```python
breast1['label'].value_counts()
```

```python
dataset=breast1
```

### VISUALIZATION

```python
plt.figure(figsize=(30,30))
sns.heatmap(dataset.corr(),annot=True,cmap='twilight_shifted')
```

```python
sns.countplot(data=dataset,x='label')
```

```python
dataset.corr()
```

### DIVIDING DATASET 

```python
x=dataset.drop(columns=['label','worst concave points'],axis=1)
x
```

```python
y=dataset['label']
y
```

### STANDARD SCALER

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
x=ss.fit_transform(x)
x
```

### SPLITING DATA

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=21)
```

### LOGISTICREGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=1000)
```

### TRAINING MODEL

```python
model1=lr.fit(x_train,y_train)
```

### PREDICTING VALUES

y1_pred=model1.predict(x_test)
y1_pred

### ACCURACY SCORE

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report,roc_curve,auc
```

```python
accuracy=100*accuracy_score(y1_pred,y_test)
accuracy
```

### DESICION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=10,criterion='entropy')
```

### TRAINING MODEL

```python
tree.fit(x_train, y_train)
```

### PREDICTING 

```python
y2_pred=tree.predict(x_test)
y2_pred
```

### ACCURACY SCORE

```python
accuracy=100*accuracy_score(y2_pred,y_test)
accuracy
```

### ADABOOSTCLASSIFIER

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc= AdaBoostClassifier()
```

### TRAINING MODEL

```python
model3=abc.fit(x_train,y_train)
```

### PREDICTING

```python
y3_pred=abc.predict(x_test)
```

### ACCURACY SCORE

```python
accuracy_score(y_test,y3_pred)
```

### CLASSIFICATION REPORT

```python
print(classification_report(y3_pred,y_test))
```

### PLOT CONFUSION MATRIX

```python
plot_confusion_matrix(model3,x_train,y_train)
```

```python
fper,tper,thr=roc_curve(y3_pred,y_test)
roc_auc = auc(fper, tper)
```

```python
plt.plot(fper, tper, label='ROC curve (area = %0.3f)' % roc_auc)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
```
